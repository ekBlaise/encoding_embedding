## Learning about One Hot Encoding Technique

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
documents = [
    "I love machine learning",
    "I enjoy deep learning",
    "I dislike bugs in my code",
    "I love debugging and fixing issues",
    "I enjoy learning new programming languages"]
    
```

```python
# Make all the words lowercase and split them into tokens
tokens = [sentence.lower().split() for sentence in documents]
tokens
```

```python
# convert data above into 2D word to provide to one hot encoder
all_words = [[word] for sentence in tokens for word in sentence]
all_words
```

```python
oh_encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = oh_encoder.fit(all_words)
one_hot_encoded
```

```python
print("Categories:", oh_encoder.categories_[0])
```

```python
# Create encoded vectors for each sentence
for sentence in tokens:
    encoded_sentence = oh_encoder.transform([[word] for word in sentence])
    print("\nSentence:", sentence)
    print(encoded_sentence)
```
