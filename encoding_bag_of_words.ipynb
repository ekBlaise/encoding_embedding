```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
```

```python
documents = [
    "I love machine learning and learning new things",
    "I enjoy deep learning",
    "I dislike bugs in my code and also fixing them",
    "I love debugging and fixing issues and learning more",
    "I enjoy learning new programming languages"]
```

```python
bow_vectorizer = CountVectorizer()
bow_encoded = bow_vectorizer.fit_transform(documents)
bow_encoded
```

```python
bow_encoded.toarray()
```

```python
### Checkout the vocabulary created by the CountVectorizer
print("Vocabulary:", bow_vectorizer.vocabulary_)
```

```python
print("Vocabulary:", bow_vectorizer.get_feature_names_out())
```
