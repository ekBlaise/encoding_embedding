### SOTA LLM Embedding

Using a CPU might produce errors when trying to import using uv, revolving around the torch library. To avoid this issue, i used conda installed
1. Create conda environment.
   
- ```conda create -n encoding python=3.11```
- ```conda activate encoding```

2. Install PyTorch via conda.
   
- ```conda install pytorch==2.2.2 torchvision==0.17.2 cpuonly -c pytorch```

3. Then install other packages
   
- ```pip install -r requirements.txt```

````python
# import necessary libraries like sentence transformers, gensim, etc.
from sentence_transformers import SentenceTransformer
import gensim
````

````python
# import the model
open_source_embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
````

````python
# Write a text or simple word and get the embedding
text = "Hello, how are you?"
embedding = open_source_embedding_model.encode(text)
print(embedding)
````

````python
# check the length of the embedding or shape of the embedding
print(len(embedding))
print(embedding.shape)
````

````python
# Given a sentence, to convert to the embedding
sentence = "This is a sample sentence for embedding."
sentence_embedding = open_source_embedding_model.encode(sentence)
print(sentence_embedding)
````

````python
# now try using a complete paragraph and get the embedding
paragraph = """In the rapidly evolving landscape of artificial intelligence, 
the ability to convert text into meaningful numerical representations has become 
increasingly crucial. This process, known as text embedding, allows machines to understand and 
process human language in a way that is both efficient and effective. By leveraging advanced 
models like Sentence Transformers, we can generate high-quality embeddings that capture the 
semantic essence of text, enabling a wide range of applications from natural language processing 
to information retrieval."""

paragraph_embedding = open_source_embedding_model.encode(paragraph)
print(paragraph_embedding)
````
