### Custom Word2vec Model Training

```python
# Import libraries
import pandas as pd
import numpy as np
import os
import gensim
import nltk
from nltk import sent_tokenize, word_tokenize
from gensim.utils import simple_preprocess
```

```python
import random
np.random.seed(42)
random.seed(42)
```

```python
# Load nltk resources for handling tokenization and other text processing tasks. 
# This is a one-time setup step.
nltk.download('all')
```

```python
# Check files to use as data for training
for file in os.listdir('./data/'):
    print(file)
```

```python
# Open one of the files to use
f = open(r"./data/story1.txt", encoding='unicode_escape')
```

```python
# Keep data in a variable
corpus = f.read()
corpus
```

```python
# Convert data to sentences
raw_sentences = sent_tokenize(corpus)
raw_sentences
```

```python
# Convert sentences into tokens
for i in raw_sentences:
    print(simple_preprocess(i))
```

```python
# Out everything together to easily get tokens
# this get the data from both files and convert them into tokens

folder_path = r"./data/"
story = []
for file in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file)
    with open(file_path, encoding='unicode_escape') as f:
        corpus = f.read()
        raw_sentences = sent_tokenize(corpus)
        for sentence in raw_sentences:
            tokens = simple_preprocess(sentence)
            story.append(tokens)
```

```python
story
```

```python
story[0]
```

```python
story[0:5]
```

```python
# Train the model
model = gensim.models.Word2Vec(
    story, 
    vector_size=100, 
    window=5, 
    min_count=2, 
    workers=4,
    sg=1           # Skip-gram model (sg=1) is used for training and sg=0 for CBOW, Default is 0
    )
```

```python
# Build the vocabulary and train the model
model.build_vocab(story)
model.train(story, total_examples=model.corpus_count, epochs=10)
```
