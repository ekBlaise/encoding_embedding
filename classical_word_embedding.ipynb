## Classical word Embedding - Word2Vec

```python
from gensim.models import Word2Vec
import gensim.downloader as api
import numpy as np
```

##### the model we are loading is trained on google news dataset and has 300 dimensions for each word vector. It contains 3 million word vectors for 3 million words and phrases.

```python
model = api.load("word2vec-google-news-300")
```

##### the above model was trained with the skip-gram architecture, which predicts the context words given a target word. The model learns to capture the semantic relationships between words based on their co-occurrence in the training data.

##### Now, to represent a sentence as a vector into the model, follow these steps:
1. Tokenize the sentence into individual words.
2. For each word in the sentence, retrieve its corresponding word vector from the model.
3. Aggregate the word vectors to create a single vector representation for the entire sentence. Common aggregation methods include averaging the word vectors or using a weighted sum based on the importance of each word in the sentence.
4. The resulting vector can then be used for various downstream tasks, such as similarity comparison, classification, or clustering.

```python
sentence = "The cat is on the roof"
word_vec = []
tokens = sentence.lower().split()
for token in tokens:
    if token in model:
        word_vec.append(model[token])
    else:
        print(f"Word '{token}' not in vocabulary.")
        
```

```python
# To create a single vector representation for the entire sentence, we can average the word vectors:
sentence_vec = np.mean(word_vec, axis=0)
```
